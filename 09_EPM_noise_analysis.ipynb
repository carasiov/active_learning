```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./data/raw/input_raw.csv", parse_dates=['date'], index_col=[0])
df = df[(df.SystemType=='SEMPRA-MICA') & (df.Serial>179800)& (df.Serial<180100)].copy()
```

```python
df_clean=df[(df['Magnet Pressure']>15.499) & (df['Magnet Pressure']<15.508) & (df['Cold Head Temperature']>27) & (df['Cold Head Temperature']<36)]
```

```python
import matplotlib.pyplot as plt

for i, serial in enumerate([179858]):  # , 179858, 180086, 180080, 179905, 179814
    group = df[df.Serial == serial]
    group_ = df_clean[df_clean.Serial == serial]

    # Erstelle eine Figur mit 3 Subplots (3 Zeilen, 1 Spalte)
    fig, axes = plt.subplots(3, 1, figsize=(10, 12), sharex=True)
    
    # Erster Plot
    axes[0].plot(group['date'], group['Pressure Heater Average Power'], label='PHAP', color='tab:blue')
    axes[0].plot(group_['date'], group_['Pressure Heater Average Power'], label='PHAP_clean', color='tab:red')
    axes[0].set_ylabel('PHAP')
    axes[0].legend()
    axes[0].grid()

    # Zweiter Plot
    axes[1].plot(group['date'], group['Magnet Pressure'], label='Pressure', color='tab:blue')
    axes[1].plot(group_['date'], group_['Magnet Pressure'], label='Pressure_clean', color='tab:red')
    axes[1].set_ylabel('Magnet Pressure')
    axes[1].legend()
    axes[1].grid()

    # Dritter Plot
    axes[2].plot(group['date'], group['Cold Head Temperature'], label='Cold Head Temperature', color='tab:blue')
    axes[2].plot(group_['date'], group_['Cold Head Temperature'], label='Cold Head Temperature_clean', color='tab:red')
    axes[2].set_ylabel('Cold Head Temp')
    axes[2].set_xlabel('Date')
    axes[2].legend()
    axes[2].grid()

    # Titel und Layout anpassen
    fig.suptitle(f'Serial {serial}', fontsize=16)
    plt.tight_layout()
    plt.show()
```
